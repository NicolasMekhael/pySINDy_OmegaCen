```python
def preprocessing(file_name,spline_smoothing_factor,periodic_spline_smoothing_factor,
                  deleted_percentile,second_deleted_percentile,n_extension,n_extension_sindy,
                  range_curve,silent=False):
    
    df=pd.read_csv('/Users/nicolasmekhael/Desktop/Summer_Project/Data/omegacen_lcvs/'+ file_name
                  ,skiprows=1, header=None,sep='   ',engine="python")    #Converting file into pandas dataframe 

    df.drop(0,axis=1, inplace=True) #Dropping collumns that are not x or t
    df.drop(3,axis=1, inplace=True)

    df.columns = ["t", "x"]
    df.sort_values("t", inplace=True)
    df.drop_duplicates(subset=['t'], keep='first', inplace=True) #Naming collumns, sorting by t and dropping dupplicates

    t=np.array(df["t"])
    x=np.array(df["x"]) #From dataframe to numpy array
    x=x-np.mean(x)

    original_data_t=t
    original_data_x=x

    ts = np.linspace(range_curve[0],range_curve[1] , 1000) #Computing spline from t and x taken from dataframe
    spl = UnivariateSpline(t, x)
    spl.set_smoothing_factor(spline_smoothing_factor)
    spline_x=spl(ts)

    if silent==False:
        plt.figure()           #Graphing spline on dataset
        plt.scatter(t,x,s=1,color="black")
        plt.plot(ts, spline_x, color='red')
        plt.xlabel("t")
        plt.ylabel("x")
        plt.title("Spline Fit with all the Data "+file_name)

    clf = LocalOutlierFactor(n_neighbors=20) #Deleting outliers in dataframe
    outlier_score=clf.fit_predict(df)
    df=df[clf.negative_outlier_factor_>np.quantile(clf.negative_outlier_factor_,deleted_percentile)].copy()

    outlier_score=clf.fit_predict(df) #Second delete for clouds of points far from curve
    df=df[clf.negative_outlier_factor_>np.quantile(clf.negative_outlier_factor_,second_deleted_percentile)].copy()

    t=np.array(df["t"]) #From dataframe (without outliers) to numpy array
    x=np.array(df["x"])
    x=x-np.mean(x)

    spl = UnivariateSpline(t, x)     #Computing spline from t and x taken from dataframe (without outliers)
    spl.set_smoothing_factor(spline_smoothing_factor)
    spline_x=spl(ts)

    if silent==False:
        plt.figure()          #Graphing spline on dataset (without outliers)
        plt.scatter(t,x,s=1,color="black")
        plt.plot(ts, spline_x, color='red')
        plt.xlabel("t")
        plt.ylabel("x")
        plt.title("Spline Fit without Outliers, Deleted Percentiles="+str(deleted_percentile)+
                  " and "+str(second_deleted_percentile)+" "+file_name)

    n_phase=2*n_extension+1
    n_phase_sindy=2*n_extension_sindy+1
    x=spline_x
    t=ts

    list_t_minus=[]
    list_t_plus=[]
    for i in range(n_extension):
        phase_delay=i+1
        list_t_minus=np.append(list_t_minus, t-n_extension+i)
        list_t_plus=np.append(list_t_plus, t+phase_delay)
    all_t=np.concatenate((list_t_minus,t,list_t_plus))
    all_x=np.tile(x, n_phase)
    all_ts=np.linspace(-n_extension,n_extension+1,n_phase*1000)

    spl = UnivariateSpline(all_t, all_x) #Compute spline over n_phase periods
    spl.set_smoothing_factor(periodic_spline_smoothing_factor)
    spline_x=spl(all_ts)    #Points of spline over n_phase periods

    ts =np.linspace(-n_extension_sindy,n_extension_sindy+1,n_phase_sindy*1000) #Time of n_phase_sindy periods
    spline_sindy=spl(ts) #Points of spline over n_phase_sindy periods

    if silent==False:
        plt.figure()    #Plotting the spline over periodic data
        plt.scatter(original_data_t,original_data_x,s=1,color="purple",label="Experimental Data")
        plt.scatter(all_t,all_x,s=1)
        plt.plot(all_ts,spline_x, color="red", label="Spline of " +str(n_phase)+ " Periods")
        plt.plot(ts,spline_sindy, color="black", label="Subset of the Spline Used for Model ("+str(n_phase_sindy)+" Periods)")
        plt.legend()
        plt.xlabel("t")
        plt.ylabel("x")
        plt.title("Getting a Truly Periodic Spline from Data without Outliers "+file_name)
    
    X = np.stack((all_ts,spline_x), axis=-1)  #Taking the derivative of the spline over n_phase periods
    sd = ps.FiniteDifference(order=2)
    xdot=sd._differentiate(X, all_ts)
    xdot=xdot[:,1]

    #Taking the subset of derivatives fed to SINDy/DMD algorithm
    Xdot = np.stack((all_ts,xdot), axis=-1)
    xdot_sindy=[]
    for i in range(len(Xdot)):
        Xdot_sublist=Xdot[i]
        if Xdot_sublist[0]>=-n_extension_sindy and Xdot_sublist[0]<=n_extension_sindy+1:
            xdot_sindy=np.append(xdot_sindy, Xdot_sublist[1])

    if silent==False:
        plt.figure()  #Plotting the derivative of the spline over periodic data
        plt.plot(all_ts,xdot, color="red", linestyle="solid",label="Derivative of the Spline Over " +str(n_phase)+ " Periods")
        plt.plot(ts,xdot_sindy, color="black", label="Subset of the Derivative Used for Model ("+str(n_phase_sindy)+" Periods)")
        plt.legend()
        plt.xlabel("t")
        plt.ylabel("xdot")
        plt.title("Derivative of the Spline without Outliers "+file_name)

    Xt=pd.DataFrame(data={'t': ts, 'x': spline_sindy, 'xdot': xdot_sindy})

    return(Xt,original_data_t,original_data_x)

def meshgrid(Xt1,Xt2):
    x1 = np.array(Xt1[['x']])
    x2 = np.array(Xt2[['x']])
    print(x1)
    print(x2)
    x1 = x1.reshape((x1.shape[0],))
    x2 = x2.reshape((x2.shape[0],))
    print(x1)
    print(x2)
    x1grid, x2grid = np.meshgrid(x1, x2)

    time = np.array(Xt[['t']])
    print(time)
    time = time.reshape((time.shape[0],))
    print(time)
```
