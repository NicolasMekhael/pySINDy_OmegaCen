```python
def SINDY_fit(file_name, spline_hyperparameter, alpha):
    df=pd.read_csv('/Users/nicolasmekhael/Desktop/Summer Project/Data/omegacen_lcvs/'+ file_name 
                  ,skiprows=1, header=None,sep='   ',engine="python")     
    
    df.drop(0,axis=1, inplace=True)
    df.drop(3,axis=1, inplace=True)

    df.columns = ["t", "x"]
    df.sort_values("t", inplace=True)
    df.drop_duplicates(subset=['t'], keep='first', inplace=True)
    
    t=np.array(df["t"])
    x=np.array(df["x"])
    
    plt.figure()
    plt.scatter(t,x,s=1,color="black")
    plt.xlabel("t")
    plt.ylabel("x")
    plt.title("Light Curve")
    
    clf = LocalOutlierFactor(n_neighbors=20)
    outlier_score=clf.fit_predict(df)
    df=df[clf.negative_outlier_factor_>np.quantile(clf.negative_outlier_factor_,0.5)].copy()
    print(clf.negative_outlier_factor_)
    #df=df[outlier_score==1].copy()
    
    
    
    t=np.array(df["t"])
    x=np.array(df["x"])
    

    plt.figure()
    plt.scatter(t,x,s=1,color="black")
    plt.xlabel("t")
    plt.ylabel("x")
    plt.title("Light Curve")

    sd_noisy = ps.SINDyDerivative(kind='spline', s=spline_hyperparameter)
    xdot=sd_noisy._differentiate(x, t)

    plt.figure()
    plt.plot(t,xdot, color="black", label="Derivative from spline of noisy data", linestyle="solid")
    plt.xlabel("t")
    plt.ylabel("xdot")
    plt.title("Derivatives")
    
    lasso_optimizer=Lasso(alpha = alpha, max_iter = 100000)

    X = np.stack((x, xdot), axis=-1)
    #sd=ps.FiniteDifference()                             #Should the differentiation method for xdot and the model be different (sd_noisy vs sd), if so, which one method for model?
    sd = ps.SINDyDerivative(kind='spline', s=5)
    
    model = ps.SINDy(feature_names=["x", "xdot"],differentiation_method=sd_noisy, feature_library=ps.PolynomialLibrary(include_interaction=True, degree=2), optimizer=lasso_optimizer)
    model.fit(X, t, unbias=True)

    model.print()
    s = model.simulate(X[0,:], t, stop_condition=None)
    nMSE = np.mean((X[:,0] - s[:,0])*(X[:,0] - s[:,0]))/np.mean((X[:,0])*(X[:,0]))
    print("nMSE: "+str(nMSE))

    plt.figure()
    plt.scatter(t,x,s=1,color="black", label="Data Points from Sky Survey")
    plt.plot(t, s[:,0], color="gray",linestyle="solid", label="Light Curve from Noisy Data with SINDY")
    plt.xlabel("t")
    plt.ylabel("x")
    plt.title("Light Curve with SINDY fit")
    plt.legend()
    plt.figure()
```

```python
def SINDY_fit_nMSE(file_name, spline_hyperparameter, alpha):
    df=pd.read_csv('/Users/nicolasmekhael/Desktop/Summer Project/Data/omegacen_lcvs/'+ file_name 
                  ,skiprows=1, header=None,sep='   ',engine="python")     
    
    df.drop(0,axis=1, inplace=True)
    df.drop(3,axis=1, inplace=True)

    df.columns = ["t", "x"]
    df.sort_values("t", inplace=True)
    df.drop_duplicates(subset=['t'], keep='first', inplace=True)

    t=np.array(df["t"])
    x=np.array(df["x"])

    sd_noisy = ps.SINDyDerivative(kind='spline', s=spline_hyperparameter)
    xdot=sd_noisy._differentiate(x, t)
    
    lasso_optimizer=Lasso(alpha = alpha, max_iter = 100000)

    X = np.stack((x, xdot), axis=-1)
    sd=ps.FiniteDifference()                           #Should the differentiation method for xdot and the model be different (sd_noisy vs sd), if so, which one method for model?
    
    model = ps.SINDy(feature_names=["x", "xdot"],differentiation_method=sd, feature_library=ps.PolynomialLibrary(include_interaction=True, degree=2), optimizer=lasso_optimizer)
    model.fit(X, t, unbias=True)

    s = model.simulate(X[0,:], t, stop_condition=None)
    nMSE = np.mean((X[:,0] - s[:,0])*(X[:,0] - s[:,0]))/(np.mean((X[:,0])*(X[:,0])))
    return(nMSE)
```

```python
def optimal_spline(file_name,alpha): #Always picks out the smallest s and models the noise (supposed to pick the best fit)
    nMSE_list=[]
    s_range=np.arange(1,5,0.02)
    for i in s_range:
        nMSE=SINDY_fit_nMSE(file_name,i,alpha)
        nMSE_list.append(nMSE)
    nMSE_list=np.array(nMSE_list)
    print("s="+str(s_range[nMSE_list.argmin()]))
    print(nMSE_list)
    SINDY_fit(file_name, s_range[nMSE_list.argmin()], alpha)
```

```python
def spline_and_outliers(file_name, deleted_percentile,spline_smoothing_factor):
    df=pd.read_csv('/Users/nicolasmekhael/Desktop/Summer Project/Data/omegacen_lcvs/'+ file_name 
                  ,skiprows=1, header=None,sep='   ',engine="python")     
    
    df.drop(0,axis=1, inplace=True)
    df.drop(3,axis=1, inplace=True)

    df.columns = ["t", "x"]
    df.sort_values("t", inplace=True)
    df.drop_duplicates(subset=['t'], keep='first', inplace=True)
    
    t=np.array(df["t"])
    x=np.array(df["x"])
    
    ts = np.linspace(0, 1, 1000)
    spl = UnivariateSpline(t, x,k=5)
    spl.set_smoothing_factor(spline_smoothing_factor)
    spline_x=spl(ts)
    
    plt.figure()
    plt.scatter(t,x,s=1,color="black")
    plt.plot(ts, spl(ts), color='red')
    plt.xlabel("t")
    plt.ylabel("x")
    plt.title("Spline Fit with all the Data")
    
    clf = LocalOutlierFactor(n_neighbors=20)
    outlier_score=clf.fit_predict(df)
    df=df[clf.negative_outlier_factor_>np.quantile(clf.negative_outlier_factor_,deleted_percentile)].copy()
    
    t=np.array(df["t"])
    x=np.array(df["x"])
    
    spl = UnivariateSpline(t, x,k=5)
    spl.set_smoothing_factor(spline_smoothing_factor)
    plt.figure()
    plt.scatter(t,x,s=1,color="black")
    plt.plot(ts, spl(ts), color='red')
    plt.xlabel("t")
    plt.ylabel("x")
    plt.title("Spline Fit without Outliers, Deleted Percentile="+str(deleted_percentile))
    
    return(t,x,ts,spline_x)
```

```python
def SINDy_fit(file_name, deleted_percentile ,spline_smoothing_factor,degree,alpha=1):

    t,x,ts,spline_x=spline_and_outliers(file_name,deleted_percentile,spline_smoothing_factor)
    
    sd = ps.SINDyDerivative(kind='spline', s=1)
    xdot=sd._differentiate(spline_x, ts)

    plt.figure()
    plt.plot(ts,xdot, color="black", label="Derivative of the Spline without Outliers", linestyle="solid")
    plt.xlabel("t")
    plt.ylabel("xdot")
    plt.title("Derivatives")
    
    lasso_optimizer=Lasso(alpha = alpha, max_iter = 100000)

    X = np.stack((spline_x, xdot), axis=-1)
    
    model = ps.SINDy(feature_names=["x", "xdot"], feature_library=ps.PolynomialLibrary(include_interaction=True, degree=degree), optimizer=lasso_optimizer)
    model.fit(X, ts, unbias=True)

    model.print()
    s = model.simulate(X[0,:], ts, stop_condition=None)
    nMSE = np.mean((X[:,0] - s[:,0])*(X[:,0] - s[:,0]))/np.mean((X[:,0])*(X[:,0]))
    print("nMSE: "+str(nMSE))

    plt.figure()
    plt.scatter(t,x,s=1,color="black", label="Data Points from Sky Survey")
    plt.plot(ts, s[:,0], color="gray",linestyle="solid", label="Light Curve from Noisy Data with SINDY")
    plt.xlabel("t")
    plt.ylabel("x")
    plt.title("Light Curve with SINDY fit")
    plt.legend()
    plt.figure()
    
SINDy_fit(os.listdir(dir_path)[266],0.2,2,2)
```

```python
DataColor = "#222"
SimulationColor = "#58008D"

def sindyfy_lasso1(star, includeInt, alpha, degree, nudge, spline_smoothing_factor, lookingForCaseStudies=False, silent=False):
    
    lasso_optimizer = Lasso(alpha = alpha, max_iter = 10000, fit_intercept=False)
    rrderiv = pd.read_csv(star)
    rrderiv.sort_values(by='Phase',inplace=True)
    rrderiv.drop_duplicates(subset=['Phase'], keep='first', inplace=True)
    print(rrderiv.head())
    Xt = rrderiv[['Phase', 'Mag']]
    t = np.array(Xt[['Phase']])
    x = np.array(Xt[['Mag']])
    spl = UnivariateSpline(t, x,k=5)
    spl.set_smoothing_factor(spline_smoothing_factor)
    spline_x=spl(t)
    
    plt.figure()
    plt.scatter(t,x,s=1,color="black")
    plt.plot(t, spl(t), color='red')
    plt.xlabel("t")
    plt.ylabel("x")
    plt.title("Spline Fit with all the Data")
    
    X = np.stack((t,spline_x), axis=-1)
    sd = ps.FiniteDifference(order=2)
    xdot=sd._differentiate(X, t)
    
    xdot=list(chain.from_iterable(list(chain.from_iterable(xdot))))
    del xdot[0:len(xdot):2]
   
    rrderiv.drop('Mag', inplace=True, axis=1)
    rrderiv.insert(0,'MagDif',xdot)
    rrderiv.insert(0,'Mag',spl(t))
    
    Xt = rrderiv[['Phase', 'Mag', 'MagDif']]
    Xt = Xt.sort_values(by='Phase')
    t = np.array(Xt[['Phase']])
    X = np.array(Xt[['Mag', 'MagDif']])
    t = t.reshape((t.shape[0],))
    t_e = 0 #round(5*len(t)/10)
    model = ps.SINDy(feature_names=["x", "xdot"],differentiation_method=ps.SINDyDerivative(kind='spline', s=2.1),feature_library=ps.PolynomialLibrary(include_interaction=includeInt, degree=degree), optimizer=lasso_optimizer)
    model.fit(X, t=t, unbias=True)
    s = model.simulate(X[0,:] + nudge, t)
    nMSE = np.mean((X[t_e:,0] - s[t_e:,0])*(X[t_e:,0] - s[t_e:,0]))/np.mean((X[t_e:,0])*(X[t_e:,0])) #MSE / MSE of predicting 0
    if(lookingForCaseStudies and nMSE > 0.2): return(model, nMSE)
    if(not silent):
        plt.figure()
        plt.scatter(t, X[:,0], s=0.2, c=DataColor)
        plt.scatter(t, s[:,0], s=0.2, c=SimulationColor)
        plt.xlabel("time")
        plt.ylabel("Magnitude (median subtracted)")
        plt.show()
        plt.scatter(t, X[:,1], s=0.2, c=DataColor)
        plt.scatter(t, s[:,1], s=0.2, c=SimulationColor)
        plt.xlabel("time")
        plt.ylabel("Magnitude derivative")
        plt.show()
        plt.scatter(X[:,0], X[:,1], s=0.2, c=DataColor)
        plt.scatter(s[:,0], s[:,1], s=0.2, c=t)
        plt.colorbar()
        plt.xlabel("Magnitude (median subtracted)")
        plt.ylabel("Magnitude derivative")
        plt.show()
        print("Governing equations:")
        model.print()
        print("Initial conditions:")
        print(X[0,:])
    return(model, nMSE)

sindyfy_lasso1("3023101028508.csv",True,1,2,0,5)
```

```python
def preprocessing(file_name,deleted_percentile,spline_smoothing_factor):
    
    t,x,ts,spline_x=spline_and_outliers(file_name,deleted_percentile,spline_smoothing_factor)
    
    X = np.stack((ts,spline_x), axis=-1)
    sd = ps.FiniteDifference(order=2)
    xdot=sd._differentiate(X, ts)
    xdot=xdot[:,1]
    
    plt.figure()
    plt.plot(ts,xdot, color="black", linestyle="solid")
    plt.xlabel("t")
    plt.ylabel("xdot")
    plt.title("Derivative of the Spline without Outliers")
    
    d={'t': ts, 'x': spline_x, 'xdot': xdot}
    Xt=pd.DataFrame(data=d)
    return(Xt)

preprocessing(os.listdir(dir_path)[784],0.1,10)
```

```python
DataColor = "#222"
SimulationColor = "#58008D"

def sindyfy_lasso(Xt, includeInt, alpha, degree, nudge, max_iter, lookingForCaseStudies=False, silent=False):
    
    lasso_optimizer = Lasso(alpha = alpha, max_iter = max_iter, fit_intercept=False)
    
    t = np.array(Xt[['t']])
    X = np.array(Xt[['x', 'xdot']])
    t = t.reshape((t.shape[0],))
    t_e = 0 #round(5*len(t)/10)
    model = ps.SINDy(feature_names=["x", "xdot"], feature_library=ps.PolynomialLibrary(include_interaction=includeInt, degree=degree), optimizer=lasso_optimizer)
    model.fit(X, t=t, unbias=True)
    s = model.simulate(X[0,:] + nudge, t)
    nMSE = np.mean((X[t_e:,0] - s[t_e:,0])*(X[t_e:,0] - s[t_e:,0]))/np.mean((X[t_e:,0])*(X[t_e:,0])) #MSE / MSE of predicting 0
    if(lookingForCaseStudies and nMSE > 0.2): return(model, nMSE)
    if(not silent):
        plt.figure()
        plt.scatter(t, X[:,0], s=0.2, c=DataColor)
        plt.scatter(t, s[:,0], s=0.2, c=SimulationColor)
        plt.xlabel("time")
        plt.ylabel("Magnitude")
        plt.show()
        plt.scatter(t, X[:,1], s=0.2, c=DataColor)
        plt.scatter(t, s[:,1], s=0.2, c=SimulationColor)
        plt.xlabel("time")
        plt.ylabel("Magnitude derivative")
        plt.show()
        plt.scatter(X[:,0], X[:,1], s=0.2, c=DataColor)
        plt.scatter(s[:,0], s[:,1], s=0.2, c=t)
        plt.colorbar()
        plt.xlabel("Magnitude")
        plt.ylabel("Magnitude derivative")
        plt.show()
        print("Governing equations:")
        model.print()
        print("Initial conditions:")
        print(X[0,:])
    return(model, nMSE)
```

```python
def results(file_name,deleted_percentile,spline_smoothing_factor, alpha, degree, nudge, max_iter=100000, includeInt=True,lookingForCaseStudies=False, silent=False):
    Xt=preprocessing(file_name,deleted_percentile,spline_smoothing_factor)
    model,nMSE=sindyfy_lasso(Xt, includeInt, alpha, degree, nudge, max_iter)
    print(nMSE)
```
